```python
import pandas as pd
import numpy as np
import keras
```

```python
data = pd.read_csv(r'E:\ESD-USB\windows\fraudTrain.csv')
```

```python
data.head(10)
```

```python
fraud_indices = np.array(data[data.is_fraud == 1].index)
number_records_fraud = len(fraud_indices)
print(number_records_fraud)
```

```python
data.shape
```

```python
pd.set_option('display.max_columns', None)
```

```python
data['trans_date_trans_time'] = pd.to_datetime(data['trans_date_trans_time'])
```

```python
data.head(10)
```

```python
data.drop(columns=['Unnamed: 0','street','state','first','last','trans_num','unix_time','gender'],inplace=True)
```

```python
data.info()
```

```python
float64_cols = list(data.select_dtypes(include='float64'))
data[float64_cols] = data[float64_cols].astype('float16')
data['cc_num']=data['cc_num'].astype('float64')
```

```python
data.info()
```

```python
data['merchant'] = data['merchant'].apply(lambda x : x.replace('fraud_',''))
```

```python
data['dob'] = pd.to_datetime(data['dob'])
data['age_cust'] = (data['trans_date_trans_time'].dt.year - data['dob'].dt.year).astype(int)
data.drop(columns='dob',inplace=True)
data.head(10)
```

```python
data['lat_dist'] = abs(round(data['merch_lat']-data['lat'],2))
data['long_dist'] = abs(round(data['merch_long']-data['long'],2))
```

```python
data.head(10)
```

```python
data.drop(columns=['lat','long','merch_lat','merch_long','trans_date_trans_time'],inplace=True)
data.shape
```

```python
data.head(10)
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
```

```python
data.drop(columns=['merchant','category','city','job','cc_num'],inplace=True)
```

```python
data.describe()
```

```python
x = data.drop('is_fraud',axis = 1)
y = data['is_fraud']
print(x[:30])
data.head(10)
```

```python
X, Y = oversample.fit_resample(x, y.values.ravel())
```

```python
print(len(X))
print(len(Y))
```

```python
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train).reshape((-1,1))
y_test = np.array(y_test).reshape((-1,1))
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train[0])
```

```python
model = Sequential([
    Dense(units=16, input_dim = 6,activation='relu'),
    Dense(units=24,activation='relu'),
    Dropout(0.5),
    Dense(20,activation='relu'),
    Dense(24,activation='relu'),
    Dense(1,activation='sigmoid'),
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,batch_size=40,epochs=8)
```

```python
score = model.evaluate(x_test, y_test)
```

```python
print(score)
```

```python
model.save("cc_fraud.h5")
```
